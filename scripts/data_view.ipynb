```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
data = pd.read_csv('../data/sampledata.csv', index_col=0).values
data.shape
z, x, y = data[:, :3].T
```

```python
fig1, ax1 = plt.subplots(3, 1, sharex=True)
ax1[0].plot(z, 'go--')
ax1[1].plot(x, 'ro--')
ax1[2].plot(y, 'bo--')

ax1[0].set_xlim(0, 100)
ax1[-1].set_xlabel(r'$t$ (step)')
fig1.suptitle("Time series from the coupled logistic map system")
```
