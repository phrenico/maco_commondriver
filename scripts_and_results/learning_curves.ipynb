# Visualization of the learning curves for the models

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

%matplotlib notebook
```

```python
# Read in results
learnings = np.load('./resdata/learning_curves.npy')
print('(epochs, restarts):', learnings.shape)
```

```python
# Clustering with k-means
nc = 2
km = KMeans(n_clusters=nc, random_state=2)
clusts = km.fit_predict(learnings[-1:, :].T)
print('# cluster1, #cluster2:', np.sum(clusts==0), np.sum(clusts==1))
```

```python
# Plot the results 
cols = ['orange', 'b']

for i in range(nc):
    _ = plt.plot(learnings[:, i==clusts], color=cols[i])

plt.xscale('log')
plt.yscale('log')

plt.ylim(0, 0.3)
plt.xlim(1, 4000)
plt.xlabel('epochs')
plt.ylabel('RMSE')
```
